```python
import numpy as np
from sklearn.decomposition import non_negative_factorization as nmf

X = np.array([ [5, 2, 4, 6], [3, 4, 6, 2]]).T
print(X)

W, H, n_iter = nmf(X)

print('W:')
print(W)
print('H:')
print(H)

print(np.dot(W,H))
```

```python
n_samples = 1000
timeframe = np.linspace(0, 50, n_samples)
print(timeframe)
```

```python
s1 = np.sin(timeframe) + 1
import matplotlib.pyplot as plt
plt.plot(s1)
```

```python
s2 = np.sign(np.sin(timeframe*1.5)) + 1
plt.plot(s2)
```

```python
s3 = 2 * np.random.rand(n_samples)
plt.plot(s3)
```

```python
def plotMatrix(s): 
    m, n = s.shape 
    if m < n: 
        s = s.T 
    m, n = s.shape 
    plt.subplots(figsize=(8, 7)) 
    x = np.arange(m) 
    for j in range(n): 
        plt.subplot(n, 1, j + 1) 
        plt.plot(x, s[:, j]) 
```

```python
S = np.vstack([s1, s2, s3])
plotMatrix(S)
```

```python
A = np.random.rand(3, 3)
print(A)
```

```python
X = np.dot(A, S)
plotMatrix(X)
```

```python
W, H, n_iter = nmf(X, n_components=3, tol=0.000001)
```

```python
plotMatrix(H)
```

```python
from sklearn.decomposition import FastICA
ica = FastICA(n_components=3)
S_ = ica.fit_transform(X.T)
A_ = ica.mixing_
plotMatrix(S_)
```
